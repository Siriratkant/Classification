```python
df2=df.drop(['BusinessTravel', 'Department' , 'EducationField', 'Gender', 'Over18' ,'JobRole', 'MaritalStatus','NumCompaniesWorked','TotalWorkingYears','EmployeeCount','Education','StandardHours','DistanceFromHome','JobLevel','PercentSalaryHike','StockOptionLevel'], axis=1)
df2.head()
```

```python
def label2val(s):
    d = {'No': 0,
         'Yes': 1 }
    if s in d.keys():
        v = d[s]
    else:
        v = np.nan
    return v
```

```python
label2val('Yes')
```

```python
df2['Attrition'].apply(label2val)
```

```python
df2[['Attrition']].applymap(label2val)
```

```python
likert_cols = ['Attrition']
df2[likert_cols]=df2[likert_cols].applymap(label2val)
```

```python
df2.info()
```

```python
df2.dropna(inplace=True)
df2
```

```python
df2.info()
```

```python
df2['EnvironmentSatisfaction'] =df2['EnvironmentSatisfaction'] .astype(int)
df2['JobSatisfaction'] =df2['JobSatisfaction'] .astype(int)
df2['WorkLifeBalance'] =df2['WorkLifeBalance'] .astype(int)
```

```python
df2.info()
```

```python
from patsy import dmatrices
import statsmodels.api as sm
```

```python
df2.columns
```

```python
y, X = dmatrices('Attrition ~ Age + MonthlyIncome  + TrainingTimesLastYear+YearsAtCompany + YearsSinceLastPromotion +YearsWithCurrManager +EnvironmentSatisfaction +JobSatisfaction + WorkLifeBalance', data=df2, return_type = 'dataframe')
```

```python
model = sm.Logit(y, X).fit()
model.summary()
```

```python
model.params
```

```python
model.predict(X)
```

```python
model.predict(X).apply(lambda p: 0 if p < .5 else 1)
```

```python
model.pred_table()
```
